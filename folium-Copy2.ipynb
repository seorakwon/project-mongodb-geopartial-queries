```python
from pymongo import MongoClient
import folium
from folium import plugins
import pandas as pd
import json


%matplotlib inline
```

```python
def connectCollection(database, collection):
    client = MongoClient()
    db = client[database]
    coll = db[collection]
    return db, coll
```

```python
db, coll = connectCollection('companies','starbucks')
print(coll.json)

starbucks = list(coll.find({"name":"Starbucks"}))
```
