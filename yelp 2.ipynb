```python
import requests
import json
import sys
import os
from dotenv import load_dotenv
load_dotenv()
import pandas as pd
```

```python
def getyelplocation(company, city):
    load_dotenv()
    search_url = "https://api.yelp.com/v3/businesses/search"
    api_key = os.getenv("YELP_API_KEY")
    headers = {'Authorization': 'Bearer %s' % api_key}
    params = {'term': company, 'location': city, "limit": 50}
    req = requests.get(search_url, params=params, headers=headers)
    print('The status code is {}'.format(req.status_code))

    response = json.loads(req.text)

    coords = []
    names = []
    city = []
    address = []
    yelp_dict = {}
    coords_list = []
    coords_2 = []

    for i in response["businesses"]:
        coords.append(i["coordinates"])
        names.append(i["name"])
        city.append(i["location"]["city"])
        address.append

    for i in coords:
        coords_list.append([i["longitude"], i["latitude"]])

    for i in range(len(coords_list)):
        longitude = coords_list[i][0]
        latitude = coords_list[i][1]
        coords_2.append(
            {"Type": "Point", "coordinates": [longitude, latitude]})

    yelp_dict = {"Name": names, "City": city,"Geolocation": coords_2}

    return pd.DataFrame(yelp_dict)
```

```python
getyelplocation("starbucks", "san francisco")
```
